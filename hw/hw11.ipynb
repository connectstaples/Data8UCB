```python
# def row_to_array(row, features):
#     """Converts a row to an array of its features."""
#     arr = make_array()
#     for feature in features:
#         arr = np.append(arr, row.item(feature))
#     return arr

# def classify(test_row, k, train):
#     test_row_features_array = row_to_array(test_row, features)
#     distances = Table().with_columns('distance', [], 'label', []) # WRONG make_array() # must be table
#     for train_row in train.rows:
#         train_row_features_array = row_to_array(train_row, features) # use row_to_array on training data 
#         row_distance = distance(train_row_features_array, test_row_features_array)
# # row_distance = distance(np.array(train_row_features_array.item(0) - train_row_features_array.item(1)))
#         distances = distances.with_row([row_distance, train_row['label']])  # Correctly access 'label'
#  # create table store the distances array with the label 
    
#     train_with_distances = distances.sort('distance', descending=False)
#     nearest_neighbors = train_with_distances.take(np.arange(k)) # compute k for table
#     most_common_label = nearest_neighbors.group('label').sort('count', descending=True).column('label').item(0) # find label from new table with group, sort, item 
#     return most_common_label 

## CORRECT CODE 

def row_to_array(row, features):
    """Converts a row to an array of its features."""
    arr = make_array()
    for feature in features:
        arr = np.append(arr, row.item(feature))
    return arr

def classify(test_row, k, train):
    test_row_features_array = row_to_array(test_row, features)
    distances = make_array()
    for train_row in train.rows:
        train_row_features_array = row_to_array(train_row, features)
        row_distance = distance(test_row_features_array, train_row_features_array)
        distances = np.append(distances, row_distance)
    train_with_distances = train.with_column('Distance', distances)
    nearest_neighbors = train_with_distances.sort('Distance').take(np.arange(k))
    # test = train_with_distances.sort('Distance').take(np.arange(k))
    most_common_label = nearest_neighbors.group('school').sort('count', descending=True).column('school').item(0)
    return most_common_label # and test.show(30)

# Don't modify/delete the code below
first_test = classify(test.row(0), 5, train)
first_test

```

**Question 1.6.** There are 77 rows of Berkeley students and 23 rows of Stanford students in the `coordinates` table. If we used the entire `coordinates` table as the training set, what is the smallest value of k that would ensure that a k-Nearest Neighbor classifier would *always* predict Berkeley as the class? Assign the value to `k`. **(10 points)**

> **Note:** Section [17.2](https://inferentialthinking.com/chapters/17/2/Training_and_Testing.html) from the textbook may be useful here!

**Thinking**: 
- with 77 Berkeley students and 23 Stanford students, you need enough of the k nearest neighbors to always be Berkeley students.
    - That means the number of Berkeley students in the k nearest neighbors must be greater than the number of Stanford students

**In Basic Classifiers (binary data)** 
- _if your goal is to guarantee that the classifier always predicts one class_
- Finding the smallest value of k such that a k-nearest neighbors classifer will always predict our x variable compared to our y variable steps:
  
    1. K to be small enough to capture local patterns
    2. Want K to be large enough to reduice noise or outliers
    3. Tune k with validation set
    4. K must be an odd number

**Rule**: 
- The reason why odd values of k are preferred in k-NN classifiers (i.e. in binary classification problems) has to do with avoiding ties in the voting process.
    - There cant be a tie when odd. Youll always have a clear majority, and the classifier will predict the class with the most neighbors.

**Problem Set**: 

- 77 UCB
- 23 Standford
- Total = 100
- To guarantee that Berkeley always wins the vote, you want to make sure that even in the worst-case scenario (where the nearest Stanford students are as close as possible), there are still more Berkeley neighbors than Stanford.
    - i.e. worst case scenario we use 24 (1 plus 23) to 'out-vote' or out-weigh the other data
    - $\text{k = (2 * 23) + 1 = 47}$

In this class, we have taught you how to use the k-NN algorithm to classify data as one of two classes. However, much of the data you will encounter in the real world will not fall nicely into one of two categories. 

**How can we classify data with non-binary classes?** It turns out we can still use k-NN! That is, we find the distance between a point and all its neighbors, find the nearest neighbors, and take a majority vote among the neighbors to determine this point's class. 

The only difference is that now the neighboring points have more than two possible classes. This does introduce difficulty because now we have no way of guaranteeing that we will not encounter ties between classes. In the case that we do encounter a tie, we can just arbitrarily choose one of the classes.

**In fact, you don't even have to modify the code you wrote before at all to enable multi-class classification!**
