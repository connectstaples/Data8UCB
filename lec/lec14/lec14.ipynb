### Notes

**Parameter** is a number associated with population
**Statistic** is a number calculated from the sample. It can estimate a parameter
**Population** the set of units that exist in nature and the set of units the conclusions apply to
**Sample** is the set of units drawn from the population

**Less than Ideal Sampling Methods**
- Deterministic Sampling (no chance involved)
- Convenience Sample (work with first units that appear)
 
**Random Sampling**_is the bes for inference_
- Before sampling, you can calculate the chance that any given subset of units is picked


**Probability Distribution**
1. all the possible values of the quantity
2. the probability of each value occurring

**Empirical Distribution**
1. all the observed values of the quantity
2. the proportion of each value occurring


**Law of Large Numbers**
- With a large sample size, the empirical distribution of a uniform random sample resembles the distribution of the population with high probability.

### The Law of Large Numbers (LLN)

_fancy definition_

**a mathematical law that states that the average of the results obtained from a large number of independent random samples converges to the true value, if it exists.**


## Theoretical Distribution 

```python
die.hist(bins = make_array(1,2,3,4,5,6,7),unit = 'spot')
print('Each spot has an equal chance of being rolled')
```

#### Observed Distributions

```python
die.sample(5).hist(bins = make_array(1,2,3,4,5,6,7),unit = 'spot')
```

```python
def observed_die_distribution(k):
    observed_die_rolls = die.sample(k)
    observed_die_rolls.hist(bins = make_array(1,2,3,4,5,6,7),unit = 'spot')

observed_die_distribution(10)
print("This doesn't look much like the theoretical distribution!")

observed_die_distribution(100)
print('This looks closer to the theoretical distribution!')

observed_die_distribution(1000)
print('This looks much more like the theoretical distribution!')

observed_die_distribution(10000)
print('This looks a lot like the theoretical distribution!')
```
